## Leetcode

```python
import pandas as pd
from datasets import load_dataset

ds = load_dataset("greengerong/leetcode")['train']
df = pd.DataFrame.from_dict(ds)
df
```

```python
idx = 0
question = df['content'][idx]
print(question)
```

```python
# Use a pipeline as a high-level helper
import torch
from transformers import pipeline

model_directory = "/datasets/ai/deepseek/hub/models--deepseek-ai--DeepSeek-R1-Distill-Qwen-7B/snapshots/6602cadec947dbb53e64f3d8d6425320b2197247"

pipe = pipeline("text-generation", 
                model=model_directory,
                max_length=8024,
                device=0)
```

```python
messages = [
    {"role": "system", "content": "Solve in python"},
    {"role": "user", "content": question},
]

pipe(messages)
```
